### Sentiment Analysis Example

```python
from transformers import pipeline

#load the sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

#analyze sentiment of a sample text
result = sentiment_analyzer("I Love hugging face!")
print(result)
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I have been waiting for a Hugging Face course.")
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("The course is bad.")
```

```python
classifier(["I have been waiting for this course my whole life.", "I hate this so much."])
```

### Zero Shot Classification


```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

```python
from transformers import pipeline

#load the zero-shot-classification pipeline
classifier = pipeline("zero-shot-classification")

#classify a text with candidate labels
result = classifier("Hugging Face is creating a tool that democratizes AI.",
                    candidate_labels = ["education", "technology", "finance"]
              )

print(result)
```

## Text Generation Example

```python
from transformers import pipeline

#Load text-generation pipeline
text_generator = pipeline("text-generation", model="gpt2")

#Generate text based on a prompt
result = text_generator("Once upon a time,")
print(result)
```

```python
from transformers import pipeline

#Load text-generation pipeline
text_generator = pipeline("text-generation")

#Generate text based on a prompt
result = text_generator("In this course about Hugging Face, we will learn")
print(result)
```

## NAMED ENTITY RECOGNITION

```python
from transformers import pipeline

# Load the NER pipeline
ner = pipeline("ner")

# Perform NER on a sample text
result = ner("Hugging Face Inc. is a company based in New York.")
print(result)
```

## Translation Example

```python
from transformers import pipeline

# Load the translation pipeline
translator = pipeline("translation_en_to_fr")

# Translate a sample text from English to French
result = translator("Hugging Face is amazing!")
print(result)
```

## Question Answering Example

```python
from transformers import pipeline

# Load the question-answering pipeline
qa = pipeline("question-answering")

# Answer a question based on context
context = "Hugging Face Inc. is a company based in New York."
result = qa(question="Where is Hugging Face based?", context=context)
print(result)
```
